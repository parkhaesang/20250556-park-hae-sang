```python
def conditional_probability(prob_a_and_b, prob_b):
    return prob_a_and_b / prob_b

# 주어진 확률값
prob_a = 0.6      # 빵을 살 확률
prob_b = 0.4      # 우유를 살 확률
prob_a_and_b = 0.3  # 빵도 사고 우유도 살 확률

# 조건부 확률 계산
prob_a_given_b = conditional_probability(prob_a_and_b, prob_b)
print("우유를 산 경우에 빵을 살 확률:", prob_a_given_b)
```

```python
!apt-get -qq -y install fonts-nanum fonts-noto-cjk > /dev/null

import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import glob

for p in glob.glob('/usr/share/fonts/truetype/nanum/*.ttf'):
    fm.fontManager.addfont(p)
for p in glob.glob('/usr/share/fonts/opentype/noto/*.otf'):
    fm.fontManager.addfont(p)
for p in glob.glob('/usr/share/fonts/truetype/noto/*.ttf'):
    fm.fontManager.addfont(p)

families = {f.name for f in fm.fontManager.ttflist}
if 'NanumGothic' in families:
    mpl.rc('font', family='NanumGothic')
elif 'Noto Sans CJK KR' in families:
    mpl.rc('font', family='Noto Sans CJK KR')
else:
    mpl.rc('font', family='DejaVu Sans')
plt.rcParams['axes.unicode_minus'] = False

colors = ['#348ABD', '#A60628']
prior = [1/25, 24/25]
posterior = [0.158, 1 - 0.158]

x = np.arange(2)
w = 0.35

plt.figure(figsize=(8,6))
plt.bar(x - w/2, prior, width=w, color=colors[0], edgecolor='#348ABD', lw=2, alpha=0.8, label='사전확률')
plt.bar(x + w/2, posterior, width=w, color=colors[1], edgecolor='#A60628', lw=2, alpha=0.8, label='사후확률')

plt.xticks(x, ['가수', '직장인'], fontsize=13)
plt.ylabel('확률', fontsize=12)
plt.title('철수의 직업에 대한 사전확률과 사후확률', fontsize=14)
plt.ylim(0, 1.05)
plt.legend(loc='upper left', fontsize=11)
plt.tight_layout()
plt.show()
```
